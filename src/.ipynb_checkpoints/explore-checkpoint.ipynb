```python
# BLOQUE 01
# CONTEXTO + OBJETIVO
# - Reusamos el dataset procesado del proyecto anterior (diabetes) para mejorar accuracy
# - Definimos rutas estándar del proyecto para cargar datos, guardar modelo y notas
# - Aseguramos reproducibilidad con random_state en todo

# comments: Standard library
from pathlib import Path

# comments: Project paths
PROJECT_ROOT = Path(".")
DATA_PROCESSED = PROJECT_ROOT / "data" / "processed"
MODELS_DIR = PROJECT_ROOT / "models"
REPORTS_DIR = PROJECT_ROOT / "reports"

# comments: Create folders if missing
DATA_PROCESSED.mkdir(parents=True, exist_ok=True)
MODELS_DIR.mkdir(parents=True, exist_ok=True)
REPORTS_DIR.mkdir(parents=True, exist_ok=True)

# comments: Reproducibility
RANDOM_STATE = 42

DATA_PROCESSED, MODELS_DIR, REPORTS_DIR, RANDOM_STATE

# Resultado esperado:
# - Carpetas data/processed, models, reports creadas si no existían
# - RANDOM_STATE definido en 42

# Interpretación:
# - Preparas el repo como pro (orden + reproducibilidad)
# - Todo lo que venga será repetible y auditable
```

```python
# BLOQUE 02
# CARGA DEL DATASET CON RUTA ROBUSTA (DESDE explore.ipynb EN /src)
# - El notebook vive en src/explore.ipynb, así que el cwd suele ser /src (no el root)
# - Construimos la ruta usando el path del notebook (fallback a cwd) para evitar FileNotFoundError
# - Cargamos diabetes.csv y validamos que existe antes de leer

# comments: Standard library
from pathlib import Path

# comments: Third-party
import pandas as pd

# comments: Resolve base directory assuming this notebook runs from /src
SRC_DIR = Path.cwd()

# comments: If we are not inside /src, force it by locating a src folder upward
if SRC_DIR.name != "src":
    for parent in [SRC_DIR] + list(SRC_DIR.parents):
        if (parent / "src").exists():
            SRC_DIR = parent / "src"
            break

# comments: Build raw data path relative to src
DATA_RAW = SRC_DIR / "data" / "raw"
csv_path = DATA_RAW / "diabetes.csv"

# comments: Validate path
if not csv_path.exists():
    raise FileNotFoundError(f"CSV not found at: {csv_path.resolve()}")

# comments: Load dataset
df = pd.read_csv(csv_path)

df.shape, df.head()

# Resultado esperado:
# - csv_path apunta a .../17-random-forest/src/data/raw/diabetes.csv
# - df carga sin errores y muestra shape + primeras filas

# Interpretación:
# - Eliminamos el problema de rutas por ejecutar el notebook dentro de /src


```

```python
# BLOQUE 03
# DEFINICIÓN DE TARGET + TRAIN/TEST SPLIT
# - Definimos la columna target (Outcome) y separamos X/y
# - Hacemos split reproducible con stratify para mantener proporción de clases
# - Guardamos los CSV procesados en src/data/processed para continuidad

# comments: Third-party
from sklearn.model_selection import train_test_split

# comments: Target column
target_col = "Outcome"
if target_col not in df.columns:
    raise ValueError(f"Target column '{target_col}' not found. Columns: {list(df.columns)}")

# comments: Feature/target split
X = df.drop(columns=[target_col])
y = df[target_col]

# comments: Train/test split (binary classification -> stratify=y)
RANDOM_STATE = 42
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    random_state=RANDOM_STATE,
    stratify=y
)

# comments: Create processed folder
DATA_PROCESSED = SRC_DIR / "data" / "processed"
DATA_PROCESSED.mkdir(parents=True, exist_ok=True)

# comments: Save split for later reuse
X_train.to_csv(DATA_PROCESSED / "X_train.csv", index=False)
X_test.to_csv(DATA_PROCESSED / "X_test.csv", index=False)
y_train.to_frame("target").to_csv(DATA_PROCESSED / "y_train.csv", index=False)
y_test.to_frame("target").to_csv(DATA_PROCESSED / "y_test.csv", index=False)

X_train.shape, X_test.shape, y_train.value_counts(normalize=True).round(3).to_dict()

# Resultado esperado:
# - X_train/X_test y y_train/y_test creados sin errores
# - Clases balanceadas de forma similar por usar stratify
# - Archivos guardados en src/data/processed

# Interpretación:
# - Dejas listo el dataset para entrenar modelos de forma limpia
# - El split queda reproducible y reutilizable en futuros notebooks

```

```python
# BLOQUE 04
# BASELINE: RANDOM FOREST DEFAULT
# - Entrenamos un RandomForestClassifier con defaults y random_state
# - Medimos accuracy y matriz de confusión como punto de partida
# - Guardamos baseline para comparar con tuning

# comments: Third-party
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# comments: Baseline model
rf_base = RandomForestClassifier(random_state=RANDOM_STATE, n_jobs=-1)
rf_base.fit(X_train, y_train)

# comments: Baseline evaluation
y_pred_base = rf_base.predict(X_test)
acc_base = accuracy_score(y_test, y_pred_base)
cm_base = confusion_matrix(y_test, y_pred_base)
report_base = classification_report(y_test, y_pred_base)

acc_base, cm_base

# Resultado esperado:
# - acc_base calculado y cm_base generado
# - baseline listo para comparación

# Interpretación:
# - Si tuning no mejora, no vale la pena
# - La matriz de confusión  muestra el tipo de errores (FN/FP)
```

```python
# BLOQUE 05
# EXPERIMENTO 1: IMPACTO DE n_estimators (BOSQUE)
# - Probamos diferentes cantidades de árboles
# - Medimos accuracy en test para comparar estabilidad vs performance
# - Graficamos accuracy vs n_estimators

# comments: Third-party
import matplotlib.pyplot as plt

# comments: Grid for n_estimators
n_estimators_grid = [50, 100, 200, 400, 800]

results_estimators = []
for n in n_estimators_grid:
    model = RandomForestClassifier(
        random_state=RANDOM_STATE,
        n_estimators=n,
        n_jobs=-1
    )
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    results_estimators.append((n, acc))

results_estimators
```

```python
# comments: Plot n_estimators impact
x = [n for n, _ in results_estimators]
y = [acc for _, acc in results_estimators]

plt.figure(figsize=(8, 4))
plt.plot(x, y, marker="o")
plt.xlabel("n_estimators")
plt.ylabel("accuracy")
plt.title("Impacto de n_estimators en accuracy (test)")
plt.xticks(x)
plt.grid(True)
plt.show()

results_estimators

# Resultado esperado:
# - Lista (n_estimators, accuracy)
# - Curva que tiende a estabilizarse al subir n_estimators

# Interpretación:
# - Subir n_estimators reduce variabilidad y mejora robustez
# - Si accuracy no sube, al menos ganas estabilidad

```

```python
# BLOQUE 06
# EXPERIMENTO 2: IMPACTO DE max_depth (ARBOL)
# - Probamos profundidades para controlar complejidad y overfitting
# - Fijamos n_estimators para aislar el efecto de max_depth
# - Graficamos accuracy vs max_depth

# comments: Grid for max_depth
max_depth_grid = [None, 3, 5, 7, 10, 15]

results_depth = []
for d in max_depth_grid:
    model = RandomForestClassifier(
        random_state=RANDOM_STATE,
        n_estimators=400,
        max_depth=d,
        n_jobs=-1
    )
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    results_depth.append((d, acc))

results_depth

# comments: Plot max_depth impact
x_labels = ["None" if d is None else str(d) for d, _ in results_depth]
y = [acc for _, acc in results_depth]

plt.figure(figsize=(8, 4))
plt.plot(range(len(x_labels)), y, marker="o")
plt.xlabel("max_depth")
plt.ylabel("accuracy")
plt.title("Impacto de max_depth en accuracy (test)")
plt.xticks(range(len(x_labels)), x_labels)
plt.grid(True)
plt.show()

results_depth

# Resultado esperado:
# - Lista (max_depth, accuracy)
# - Un sweet spot típico en depth medio (depende del dataset)

# Interpretación:
# - Depth demasiado alto puede sobreajustar
# - Depth medio suele generalizar mejor

```

```python
# BLOQUE 07
# SELECCIÓN DEL MEJOR MODELO + EVALUACIÓN FINAL
# - Elegimos el mejor de cada experimento por accuracy
# - Definimos params finales y reentrenamos
# - Mostramos accuracy + confusion matrix + report

# comments: Pick best configurations
best_n, best_acc_n = max(results_estimators, key=lambda t: t[1])
best_d, best_acc_d = max(results_depth, key=lambda t: t[1])

best_n, best_acc_n, best_d, best_acc_d

```

```python
# comments: Choose final params (pick the higher of the two best accuracies)
if best_acc_d >= best_acc_n:
    final_params = {"n_estimators": 400, "max_depth": best_d}
else:
    final_params = {"n_estimators": best_n, "max_depth": None}

# comments: Train final model
rf_final = RandomForestClassifier(
    random_state=RANDOM_STATE,
    n_jobs=-1,
    **final_params
)
rf_final.fit(X_train, y_train)

# comments: Final evaluation
y_pred_final = rf_final.predict(X_test)
acc_final = accuracy_score(y_test, y_pred_final)
cm_final = confusion_matrix(y_test, y_pred_final)
report_final = classification_report(y_test, y_pred_final)

final_params, acc_final, cm_final

# comments: Print report
print(report_final)

# Resultado esperado:
# - acc_final >= acc_base (ideal)
# - Matriz de confusión y reporte impresos

# Interpretación:
# - Si acc_final supera baseline, tuning valió la pena
# - El report te dice si el modelo falla en detectar diabetes (recall clase 1)
```

```python
# BLOQUE 08
# GUARDADO DEL MODELO FINAL
# - Guardamos el modelo entrenado en src/models con nombre descriptivo
# - Incluimos n_estimators, max_depth y seed para trazabilidad
# - Esto permite reutilizar el modelo sin reentrenar

# comments: Standard library
from pathlib import Path
from pickle import dump

# comments: Models folder (project root is parent of src)
PROJECT_ROOT = SRC_DIR.parent
MODELS_DIR = PROJECT_ROOT / "models"
MODELS_DIR.mkdir(parents=True, exist_ok=True)

# comments: Build filename
n_str = final_params.get("n_estimators", rf_final.n_estimators)
d_val = final_params.get("max_depth", rf_final.max_depth)
d_str = "None" if d_val is None else str(d_val)

model_path = MODELS_DIR / f"random_forest_diabetes_n{n_str}_d{d_str}_seed{RANDOM_STATE}.sav"
dump(rf_final, open(model_path, "wb"))

str(model_path)

# Resultado esperado:
# - Archivo .sav guardado en /models

# Interpretación:
# - Tu entrega queda lista para uso real (deploy/reuse)
# - El nombre me permite recordar la configuración exacta

```
